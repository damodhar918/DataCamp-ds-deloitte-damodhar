```python
import pandas as pd 
import numpy as np
import sqlite3

conn = sqlite3.connect('example.db')
# df = pd.read_csv('voters.csv')
# df.to_sql('voters', conn, if_exists='replace', index=False)
def rsq(query: str) -> pd.DataFrame:
    return pd.read_sql_query(query, conn)
```

```python
rsq("""SELECT
first_name,
birthdate,
first_vote_date,
-- Select the diff between the 18th birthday and first vote
STRFTIME('%Y', DATE(first_vote_date)) - STRFTIME('%Y', DATE(birthdate)) - CASE WHEN STRFTIME('%m-%d', DATE(first_vote_date)) < STRFTIME('%m-%d', DATE(birthdate)) THEN 1 ELSE 0 END AS adult_years_until_vote,
-- Get the difference in weeks from 2019-01-01 until now
CAST((JULIANDAY('now') - JULIANDAY('2019-01-01')) / 7 AS INTEGER) AS weeks_passed
FROM voters;
""")

```
