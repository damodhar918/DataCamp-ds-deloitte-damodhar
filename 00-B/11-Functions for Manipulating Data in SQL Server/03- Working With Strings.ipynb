```python
rsq("""SELECT
(company) as company,
bean_type,
(broad_bean_origin) as broad_bean_origin,
-- 'company' and 'broad_bean_origin' should be in uppercase
'The company ' || UPPER(company) || ' uses beans of type "' || bean_type || '", originating from ' || UPPER(broad_bean_origin) || '.' as Highliter
FROM ratings
WHERE
-- The 'broad_bean_origin' should not be unknown
LOWER(broad_bean_origin) NOT LIKE '%unknown%'
-- The 'bean_type' should not be unknown
AND LOWER(bean_type) NOT LIKE '%unknown%';""")
```

```python
rsq("""

-- Query with concatenated messages
SELECT bean_type,
       bean_origin,
       cocoa_percent,
       -- Create a message by concatenating values with "||"
       (SELECT 'Chocolate with beans from' || ' ' || bean_origin || ' ' || 'has a cocoa percentage of' || ' ' || CAST(cocoa_percent AS nvarchar)) AS message1,
       -- Create a message by concatenating values with "||"
       (SELECT 'Chocolate with beans from' || ' ' || bean_origin || ' ' || 'has a cocoa percentage of' || ' ' || cocoa_percent) AS message2,
       -- Create a message by concatenating values with "||"
       (SELECT 'Chocolate with beans from' || ' ' || bean_origin || ' ' || 'has a cocoa percentage of' || ' ' || cocoa_percent ) AS message3
FROM ratings
WHERE company = 'Ambrosia'
  AND bean_type <> 'Unknown';
""")
```

```python
SELECT
company,
-- Create a list with all bean origins ordered alplabetically
STRING_AGG(bean_origin, ',') WITHIN GROUP (ORDER BY bean_origin) AS bean_origins
FROM ratings
WHERE company IN ('Bar Au Chocolat', 'Chocolate Con Amor', 'East Van Roasters')
-- Specify the columns used for grouping your data
GROUP BY company;
```

```python
rsq("""SELECT
company,
-- Create a list with all bean origins ordered alphabetically
GROUP_CONCAT(bean_origin, ',' ORDER BY bean_origin) AS bean_origins
FROM ratings
WHERE company IN ('Bar Au Chocolat', 'Chocolate Con Amor', 'East Van Roasters')
-- Specify the columns used for grouping your data
GROUP BY company;""")
```
