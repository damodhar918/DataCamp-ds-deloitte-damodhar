```python
import sqlite3
import pandas as pd
conn = sqlite3.connect(r'./example.db')
# SQL statement to drop table
# conn.execute("DROP TABLE album")
# commit changes and close connection
conn.commit()
# conn.close()
df = pd.read_csv(r'./films.csv',on_bad_lines='skip')
df.to_sql('films',conn,if_exists='replace')
df = pd.read_csv(r'./people.csv',on_bad_lines='skip')
df.to_sql('people',conn,if_exists='replace')
df = pd.read_csv(r'./reviews.csv',on_bad_lines='skip')
# /add indexer start with 1
df['id'] = pd.array(range(1,len(df)+1))
df.to_sql('reviews',conn,if_exists='replace')
df = pd.read_csv(r'./roles.csv',on_bad_lines='skip')
df.to_sql('roles',conn,if_exists='replace')
```
