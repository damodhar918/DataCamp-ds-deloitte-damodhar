```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sqlite3

conn = sqlite3.connect('example.db')


def rsq(query: str) -> pd.DataFrame:
    return pd.read_sql_query(query, conn)
```

```python
rsq("""-- Write a query that returns an aggregation 
SELECT MixDesc, SUM(Quantity) AS Total
FROM Shipments
GROUP BY MixDesc;""")
```

```python
rsq("""-- Count the number of rows by MixDesc
SELECT MixDesc, COUNT(*)
FROM Shipments
GROUP BY MixDesc""")
```

```python
rsq("""-- Return the difference in OrderDate and ShipDate
SELECT OrderDate, ShipDate, 
       DATEDIFF(day, OrderDate, ShipDate) AS Duration
FROM Shipments""")
```

```python
rsq("""-- Return the difference in OrderDate and ShipDate
SELECT OrderDate, ShipDate, 
    julianday(ShipDate) - julianday(OrderDate) AS Duration, count(*)
FROM Shipments group by 3; """)
```

```python
"""-- Return the DeliveryDate as 5 days after the ShipDate
SELECT OrderDate, 
DATEADD(day, 5, ShipDate)  AS DeliveryDate
FROM Shipments"""
```

```python
rsq("""-- Return the DeliveryDate as 5 days after the ShipDate
SELECT OrderDate, 
       date(ShipDate, '+5 day') AS DeliveryDate
FROM Shipments""")
```

```python
rsq("""-- Round Cost to the nearest dollar
SELECT Cost, 
       ROUND(Cost,0) AS RoundedCost, count(*) 
FROM Shipments group by 2;""")
```

```python
rsq("""-- Truncate cost to whole number
SELECT Cost, 
       round(cost,0,1) AS TruncateCost
FROM Shipments""")
```

```python
rsq("""-- Truncate cost to whole number
SELECT Cost, 
    round(cost,0) AS TruncateCost
FROM Shipments""")
```

```python
rsq("""-- Return the absolute value of DeliveryWeight
SELECT DeliveryWeight,
       ABS(DeliveryWeight) AS AbsoluteValue, count(*)
FROM Shipments group by 2;""")
```

```python
rsq("""-- Return the square and square root of WeightValue
SELECT WeightValue, 
       POWER(WeightValue, 2) AS WeightSquare, 
       SQRT(WeightValue) AS WeightSqrt
FROM Shipments""")
```
