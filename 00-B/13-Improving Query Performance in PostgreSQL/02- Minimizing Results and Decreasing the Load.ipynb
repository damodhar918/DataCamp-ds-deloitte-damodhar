```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sqlite3

engine = sqlite3.connect('example.db')
def rsq(quary:str) -> pd.DataFrame:
    return pd.read_sql_query(quary, engine)
```

```python
rsq(""" 
select olympic_cc, year, round(gdp, -3) as gdp_rounded, round(population, -3) as population_rounded, rank() over (partition by year order by gdp desc) as gdp_rank, rank() over (partition by year order by population desc) as population_rank
from demographics

""")
```

```python
rsq("""
SELECT 
    d.olympic_cc,
    d.year,
    ROUND(d.gdp, -3) AS gdp_rounded,
    ROUND(d.population, -3) AS population_rounded,
    CASE 
        WHEN r.gdp_rank <= (SELECT COUNT(*) FROM demographics WHERE year = d.Year)/3 THEN 'High'
        WHEN r.gdp_rank <= 2*(SELECT COUNT(*) FROM demographics WHERE year = d.Year)/3 THEN 'Medium'
        ELSE 'Low'
    END AS gdp_rank,
    CASE 
        WHEN r.population_rank <= (SELECT COUNT(*) FROM demographics WHERE year = d.Year)/3 THEN 'High'
        WHEN r.population_rank <= 2*(SELECT COUNT(*) FROM demographics WHERE year = d.Year)/3 THEN 'Medium'
        ELSE 'Low'
    END AS population_rank
FROM demographics d
JOIN (
    SELECT 
        year,
        olympic_cc,
        gdp,
        RANK() OVER (PARTITION BY year ORDER BY gdp DESC) AS gdp_rank,
        population,
        RANK() OVER (PARTITION BY year ORDER BY population DESC) AS population_rank
    FROM demographics
) r ON d.year = r.year AND d.olympic_cc = r.olympic_cc
ORDER BY d.year, d.olympic_cc;
""").to_sql("demographics_rank", engine, if_exists='replace', index=False)
```

```python
rsq("""-- South African athletes by year
with athletes_cte as 
(
    SELECT year
      , season
      , COUNT(DISTINCT athlete_id) AS no_athletes
    FROM athletes
    WHERE country_code = 'RSA' -- South Africa filter
    GROUP BY year, season
)

SELECT ath.year
  , ath.season
  , ath.no_athletes
  , demos.gdp_rounded
  , demos.gdp_rank
  , demos.population_rounded
  , demos.population_rank
FROM athletes_cte ath
INNER JOIN demographics_rank demos
  ON ath.year = demos.year
  and demos.olympic_cc = 'RSA' -- Filter to South Africa
ORDER BY ath.season, ath.year;""")
```
