Which market is better?
The key metric that the company uses to evaluate salespeople is the percent of sales they make over $1000 since the time put into each sale is usually worth a bit more than that, so the higher this metric, the better the salesperson is performing.

Recall that Amir's current sales amounts have a mean of $5000 and a standard deviation of $2000, and Amir's predicted amounts in next quarter's market have a mean of $6000 and a standard deviation of $2600.

norm from scipy.stats is imported.

Based only on the metric of percent of sales over $1000, does Amir perform better in the current market or the predicted market?


```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the parameters of the current market 
current_mean = 5000
current_std = 2000

# Define the parameters of the predicted market 
predicted_mean = 6000
predicted_std = 2600

# Define the range for the x-axis
x_axis = np.arange(0, 10001, 1)

# Calculate the probability of a sale over $1000 in the current market
current_prob = 1 - norm.cdf(1000, loc=current_mean, scale=current_std)

# Calculate the probability of a sale over $1000 in the predicted market 
predicted_prob = 1 - norm.cdf(1000, loc=predicted_mean, scale=predicted_std)

# Print the probabilities 
print("Probability of a sale over $1000 in current market:", current_prob)
print("Probability of a sale over $1000 in predicted market:", predicted_prob)

# Compare the probabilities 
if round(current_prob, 1) == round(predicted_prob, 1):
    print("Amir performs well in both markets.")
elif predicted_prob > current_prob:
    print("Amir performs better in the predicted market.")
else:
    print("Amir performs better in the current market.")

# Plot the probability density function for the current market
current_pdf = norm.pdf(x_axis, loc=current_mean, scale=current_std)
plt.plot(x_axis, current_pdf, label='Current Market PDF')

# Plot the probability density function for the predicted market
predicted_pdf = norm.pdf(x_axis, loc=predicted_mean, scale=predicted_std)
plt.plot(x_axis, predicted_pdf, label='Predicted Market PDF')

# Add legend and titles to the plot
plt.legend()
plt.title("Probability Density Function for Current and Predicted Markets")
plt.xlabel("Sale Price")
plt.ylabel("Probability Density")

# Show the plot
plt.show()
```

```python
np.random.seed(321)
all_deals = pd.read_csv('all_deals.csv')    
sample_means = []
# Loop 30 times to take 30 means
for i in range(30):
    # Take sample of size 20 from num_users col of all_deals with replacement
    cur_sample = all_deals['num_users'].sample(20, replace=True)
    # Take mean of cur_sample
    cur_mean = np.mean(cur_sample)
    # Append cur_mean to sample_means
    sample_means.append(cur_mean)


# Print the mean of sample_means
print(f"The mean of the sample means is {np.mean(sample_means)}")

# Print the mean of the num_users column in all_deals
print(f"The mean of 'num_users' column in 'all_deals' is {np.mean(all_deals['num_users'])}")

# Create a histogram of the sample means
plt.hist(sample_means)
plt.title('Distribution of Sample Means')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.axvline(x=np.mean(sample_means), color='r', label='Sample Mean')
plt.axvline(x=np.mean(all_deals['num_users']), color='g', label='Population Mean')
plt.legend()
plt.show()
```

```python
# There aren't any major issues with the code, so I will just make some minor changes to improve its readability and consistency with PEP 8 style guides.

import matplotlib.pyplot as plt 
from scipy.stats import poisson 

# Calculate the probability that Amir responds to 5 leads in a day, given that he responds to an average of 4.
prob_5 = poisson.pmf(5, 4) 
print("Probability of 5 responses:", prob_5)

# Calculate the probability that Amir's coworker responds to 5 leads in a day, given that she responds to an average of 5.5.
coworker_prob_5 = poisson.pmf(5, 5.5)
print("Probability of 5 coworker responses:", coworker_prob_5)

# Calculate the probability that Amir responds to 2 or fewer leads in a day.
prob_2_or_less = poisson.cdf(2, 4)
print("Probability of 2 or fewer responses:", prob_2_or_less)

# Calculate the probability that Amir responds to more than 10 leads in a day.
prob_over_10 = 1 - poisson.cdf(10, 4)
print("Probability of more than 10 responses:", prob_over_10)

# Plotting the PMF for lambda = 4
x = range(0, 15)
plt.plot(x, poisson.pmf(x, 4), 'bo', ms=8, label='Poisson PMF')

# Adding annotations and print points on plot
plt.vlines(5, 0, 0.2, colors='r', linestyles='--', label='5 responses')
plt.text(5.1, 0.2, "Prob: {:.4f}".format(prob_5), va='center', ha='left', color='r')
plt.vlines(2, 0, 0.07, colors='g', linestyles='--', label='2 responses or less')
plt.text(2.1, 0.07, "Prob: {:.4f}".format(prob_2_or_less), va='center', ha='left', color='g')
plt.legend(loc='best')
plt.title('Poisson PMF for $\lambda=4$')
plt.xlabel('Number of responses')
plt.ylabel('Probability')
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy.stats import expon
# Define the x-axis values
x = np.linspace(0, 10, 1000)

# Define the exponential distribution with scale=2.5
scale = 2.5
exp_dist = expon(scale=scale)

# Define the y-axis values
y = exp_dist.pdf(x)

# Define points
x_points = [1, 4, 3.5]
y_points = [exp_dist.pdf(1), exp_dist.pdf(4), exp_dist.pdf(3.5)]

# Print probability response takes < 1 hour
print(exp_dist.cdf(1))

# Print probability response takes > 4 hours
print(exp_dist.sf(4))

# Print probability response takes 3-4 hours
print(exp_dist.cdf(4) - exp_dist.cdf(3))

# Plot the exponential distribution
plt.plot(x, y)
plt.scatter(x_points, y_points, color='r')
plt.xlabel('Time (hours)')
plt.ylabel('Probability Density')
plt.title('Exponential Distribution with Scale={}'.format(scale))
plt.show()
```
