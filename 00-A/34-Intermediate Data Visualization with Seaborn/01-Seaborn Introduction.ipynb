```python
# import all modules
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

grant_file =r'https://assets.datacamp.com/production/course_7030/datasets/schoolimprovement2010grants.csv'
# Read in the DataFrame
df = pd.read_csv(grant_file,index_col=0)
df.head()
```

```python
##
# Display pandas histogram
df['Award_Amount'].plot.hist()
plt.show()

# Clear out the pandas histogram
plt.clf()

##
# Display a Seaborn displot
sns.displot(df['Award_Amount'])
plt.show()

# Clear the displot
plt.clf()
```

```python
# Create a displot
sns.displot(df['Award_Amount'],
             bins=20)

# Display the plot
plt.show()
```

```python
# Create a displot of the Award Amount
sns.displot(df['Award_Amount'],
             kind='kde',
             rug=True,
             fill=True)

# Plot the results
plt.show()
```

```python
sns.displot(df['Award_Amount'], kind='ecdf')
```

```python
df= pd.read_csv('./insurance_premiums.csv')
df.head()
```

```python
##
# Create a regression plot of premiums vs. insurance_losses
sns.regplot(data=df, x="insurance_losses", y="premiums")

# Display the plot
plt.show()
```

```python

##
# Create an lmplot of premiums vs. insurance_losses
sns.lmplot(data=df, y="premiums", x="insurance_losses")

# Display the second plot
plt.show()
```

```python
# Create a regression plot using hue
sns.lmplot(data=df,
           x="insurance_losses",
           y="premiums",
           hue="Region")

# Show the results
plt.show()
```

```python
# Create a regression plot with multiple rows
sns.lmplot(data=df,
           x="insurance_losses",
           y="premiums",
           row="Region",
           height=6,)

# Show the plot
plt.show()
```
