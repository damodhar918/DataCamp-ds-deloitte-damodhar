```python
# STEP-3 IMPORTING/READING THE DOCUMENT

def read_doc(directory):
    file_loader=PyPDFDirectoryLoader(directory)
    documents=file_loader.load()
    return documents
```

```python
doc=read_doc('Datasource/')
doc
```

```python
len(doc)
```

```python
#STEP-4 CONVERTING THE DATASOURCE INTO TEXT CHUNKS

def chunk_data(docs,chunk_size=400,chunk_overlap=20):
    text_splitter=RecursiveCharacterTextSplitter(chunk_size=chunk_size,chunk_overlap=chunk_overlap)
    doc=text_splitter.split_documents(docs)
    return doc
```

```python
documents=chunk_data(docs=doc)
documents
```

```python
documents=chunk_data(docs=doc)
len(documents)
```
