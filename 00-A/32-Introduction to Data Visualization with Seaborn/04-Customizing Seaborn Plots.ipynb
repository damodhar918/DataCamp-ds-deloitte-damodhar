```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

survey_data = pd.read_csv('./young-people-survey-responses.csv',index_col=0)
survey_data.head()
```

```python
survey_data.info()
survey_data['Parents\' advice'].unique()
```

```python
# survey_data['Parents\' advice' ] = survey_data['Parents\' advice'].astype('int')
# survey_data['Parents Advice'] = survey_data['Parents\' advice'].map(
#     {'Never': , 'Rarely': 1, 'Sometimes': 2, 'Often': 3, 'Always': 4 }
# )
survey_data['Parents Advice'] = pd.cut(survey_data['Parents\' advice'], bins=[-np.inf, 1, 2, 3, 4, np.inf], labels=['Never', 'Rarely', 'Sometimes', 'Often', 'Always'])
```

```python
# Set the color palette to "Purples"
sns.set_style("whitegrid")


# Create a count plot of survey responses
category_order = ["Never", "Rarely", "Sometimes", 
                  "Often", "Always"]

sns.catplot(x="Parents Advice", 
            data=survey_data, 
            kind="count", 
            order=category_order)

# Show plot
plt.show()
```

```python
survey_data['Siblings'].fillna(0, inplace=True)
survey_data['Siblings']  = survey_data['Siblings'].astype(int)
survey_data['Number of Siblings']  = pd.cut(survey_data['Siblings'], bins=[-np.inf, 0, 2, np.inf], labels=['0', '1-2', '3+'])
survey_data['Feels Lonely'] = np.where(survey_data['Loneliness'] > 3, True, False)
```

```python
# Change the context to "notebook"
sns.set_context("paper")

# Create bar plot
sns.catplot(x="Number of Siblings", y="Feels Lonely",
            data=survey_data, kind="bar")

# Show plot

plt.show()
```

```python
mpg = pd.read_csv('./mpg.csv')
```

```python
# Create scatter plot
g = sns.relplot(x="weight", 
                y="horsepower", 
                data=mpg,
                kind="scatter")

# Identify plot type
type_of_g = type(g)

# Print type
print(type_of_g)

# <script.py> output:
#     <class 'seaborn.axisgrid.FacetGrid'>
```

```python
# Create scatter plot
g = sns.relplot(x="weight", 
                y="horsepower", 
                data=mpg,
                kind="scatter")

# Add a title "Car Weight vs. Horsepower"
g.fig.suptitle("Car Weight vs. Horsepower")

# Show plot
plt.show()
```

```python
mpg_mean = mpg.groupby(['model_year','origin'])['mpg'].mean().reset_index()
mpg_mean.rename(columns={'mpg': 'mpg_mean'}, inplace=True)
mpg_mean.head()
sns.set_palette("Blues")
```

```python
# Create line plot
g = sns.lineplot(x="model_year", y="mpg_mean", 
                 data=mpg_mean,
                 hue="origin")

# Add a title "Average MPG Over Time"
g.set_title("Average MPG Over Time")

# Add x-axis and y-axis labels
g.set(xlabel="Car Model Year", ylabel="Average MPG")

# Show plot
plt.show()
```

```python
# Create point plot
sns.catplot(x="origin", 
            y="acceleration", 
            data=mpg, 
            kind="point", 
            # join=False, 
            color="#bcbd22", 
            linestyle='none',
            capsize=0.1)

# Rotate x-tick labels
plt.xticks(rotation=90)

# Show plot
plt.show()
```

```python
survey_data['Interested in Pets'] = np.where(survey_data['Pets'] > 3,'Yes','No')   
```
