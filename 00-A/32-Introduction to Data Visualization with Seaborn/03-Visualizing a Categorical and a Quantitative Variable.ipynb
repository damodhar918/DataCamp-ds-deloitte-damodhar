```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

survey_data = pd.read_csv('./young-people-survey-responses.csv',index_col=0)
survey_data.head()
```

```python
# Create count plot of internet usage
sns.catplot(x="Internet usage", data=survey_data,
            hue='Internet usage',label='Internet usage', kind="count")
# Show plot
plt.show()
```

```python
# Change the orientation of the plot
sns.catplot(y="Internet usage", data=survey_data,
            kind="count")

# Show plot
plt.show()
```

```python
survey_data['Age Category'] = pd.cut(survey_data['Age'], bins=[0, 21, np.inf], labels=['Less than 21', '21+'])
```

```python
# Separate into column subplots based on age category
sns.catplot(y="Internet usage", data=survey_data,
            kind="count", col="Age Category")

# Show plot
plt.show()
```

```python
survey_data['Interested in Math'] = np.where(survey_data['Mathematics']>3, True, False)    
```

```python
# Create a bar plot of interest in math, separated by gender
sns.catplot(x="Gender", y="Interested in Math", data=survey_data, kind="bar",hue='Gender')

# Show plot
plt.show()
```

```python
student_data = pd.read_csv('./student-alcohol-consumption.csv',index_col=0)
student_data.head()
```

```python
# List of categories from lowest to highest
category_order = ["<2 hours", 
                  "2 to 5 hours", 
                  "5 to 10 hours", 
                  ">10 hours"]

# Rearrange the categories
sns.catplot(x="study_time", y="G3",
            data=student_data,
            kind="bar")

# Show plot
plt.show()
```

```python
# List of categories from lowest to highest
category_order = ["<2 hours", 
                  "2 to 5 hours", 
                  "5 to 10 hours", 
                  ">10 hours"]

# Turn off the confidence intervals
sns.catplot(x="study_time", y="G3",
            data=student_data,
            kind="bar",
            order=category_order)

# Show plot
plt.show()
```

```python
# List of categories from lowest to highest
category_order = ["<2 hours", 
                  "2 to 5 hours", 
                  "5 to 10 hours", 
                  ">10 hours"]

# Turn off the confidence intervals
sns.catplot(x="study_time", y="G3",
            data=student_data,
            kind="bar",
            order=category_order, errorbar=None, hue='study_time')

# Show plot
plt.show()
```

```python
# Specify the category ordering
study_time_order = ["<2 hours", "2 to 5 hours", 
                    "5 to 10 hours", ">10 hours"]

# Create a box plot and set the order of the categories
sns.catplot(x="study_time", y="G3", data=student_data, order=study_time_order, kind="box")

# Show plot

plt.show()
```

```python
sns.catplot(data=student_data, x='internet', y='G3', kind='box', hue='location', fliersize=0)

# Set plot labels and title
plt.xlabel('Has Internet Access')
plt.ylabel('Final Grade')
plt.title('Distribution of Final Grades by Internet Access and Location')

# Show plot
plt.show()
```

```python
##
# Set the whiskers to 0.5 * IQR
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box", whis=0.5)

# Show plot
plt.show()
```

```python

##
# Extend the whiskers to the 5th and 95th percentile
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box",
            whis=[5,95])

# Show plot
plt.show()
```

```python


##
# Set the whiskers at the min and max values
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box",
            whis=[0, 100])

# Show plot
plt.show()
```

```python
# Add caps to the confidence interval
sns.catplot(x="famrel", y="absences",
			data=student_data,
            kind="point")
        
# Show plot
plt.show()
```

```python
# Remove the lines joining the points
sns.catplot(x="famrel", y="absences",
			data=student_data,
            kind="point",
            capsize=0.2)
            
# Show plot
plt.show()
```

```python
# Remove the lines joining the points
sns.catplot(x="famrel", y="absences",
			data=student_data,
            kind="point",
            capsize=0.2, linestyle='none')
            
# Show plot
plt.show()
```

```python
# Turn off the confidence intervals for this plot
sns.catplot(x="romantic", y="absences",
			data=student_data,
            kind="point",
            hue="school")

# Show plot
plt.show()
```

```python
# Import median function from numpy
from numpy import median

# Plot the median number of absences instead of the mean
sns.catplot(x="romantic", y="absences",
			data=student_data,
            kind="point",
            hue="school",
            errorbar=None)

# Show plot
plt.show()
```

```python
# Import median function from numpy
from numpy import median

# Plot the median number of absences instead of the mean
sns.catplot(x="romantic", y="absences",
			data=student_data,
            kind="point",
            hue="school",
            errorbar=None, estimator=median)

# Show plot
plt.show()
```
